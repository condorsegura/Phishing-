## Pergunta de Negócio 5:

### Qual Segmento Teve o Maior Total de Vendas?

Demonstre o resultado através de um gráfico de pizza.

```python
df_13 = df.groupby('Segmento')['Valor_Venda'].sum().reset_index().sort_values(by = 'Valor_Venda',ascending = False)
```

```python
df_13.head()
```

```python
# Função para converter os dados em valor absoluto
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return ' $ {v:d}'.format(v = val)
    return my_format
```

```python
# Plot

# Tamanho da figura
plt.figure(figsize = (16, 6))

# Gráfico de pizza
plt.pie(df_13['Valor_Venda'],labels = df_13['Segmento'],autopct = autopct_format(df_13['Valor_Venda']),startangle = 90)

# Limpa o círculo central
centre_circle = plt.Circle((0, 0), 0.82, fc = 'silver')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Labels e anotações
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(df_13['Valor_Venda']))), xy = (-0.25, 0))
plt.title('Total de Vendas Por Segmento')
plt.show()
```

## Pergunta de Negócio 7 (Desafio Nível Júnior):

Os gestores da empresa estão considerando conceder diferentes faixas de descontos e gostariam de fazer uma simulação com base na regra abaixo:

- Se o Valor_Venda for maior que 1000 recebe 15% de desconto.
- Se o Valor_Venda for menor que 1000 recebe 10% de desconto.

### Quantas Vendas Receberiam 15% de Desconto?

```python
import numpy as np
```

```python
# Cria uma nova coluna de acordo com a regra definida acima

# df['Desconto']: Cria uma nova coluna chamada Desconto' no DataFrame df.

#np.where(condição, valor_se_verdadeiro, valor_se_falso): Essa função do NumPy avalia a
#condição para cada linha do DataFrame.

#df['Valor_Venda'] > 1000: Define a condição: se o valor na coluna Valor_Venda for maior
#que 1000.
#0.15: O valor a ser atribuído à coluna Desconto' se a condição for verdadeira
 #(desconto de 15%).
#0.10: O valor a ser atribuído à coluna Desconto' se a condição for falsa (desconto de 10%).

df['Desconto'] = np.where(df['Valor_Venda'] > 1000, 0.15, 0.10)
```
