```python
# Função para converter os dados em valor absoluto
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct * total / 100.0))
        return ' $ {v:d}'.format(v = val)
    return my_format
```

```python
# Plot

# Tamanho da figura
plt.figure(figsize = (16, 6))

# Gráfico de pizza
plt.pie(df_13['Valor_Venda'],labels = df_13['Segmento'],autopct = autopct_format(df_13['Valor_Venda']),startangle = 90)

# Limpa o círculo central
centre_circle = plt.Circle((0, 0), 0.82, fc = 'silver')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Labels e anotações
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(df_13['Valor_Venda']))), xy = (-0.25, 0))
plt.title('Total de Vendas Por Segmento')
plt.show()
```

## Pergunta de Negócio 7

Os gestores da empresa estão considerando conceder diferentes faixas de descontos e gostariam de fazer uma simulação com base na regra abaixo:

- Se o Valor_Venda for maior que 1000 recebe 15% de desconto.
- Se o Valor_Venda for menor que 1000 recebe 10% de desconto.

### Quantas Vendas Receberiam 15% de Desconto?

```python
import numpy as np
```

```python
# Cria uma nova coluna de acordo com a regra definida acima

# df['Desconto']: Cria uma nova coluna chamada Desconto' no DataFrame df.

#np.where(condição, valor_se_verdadeiro, valor_se_falso): Essa função do NumPy avalia a
#condição para cada linha do DataFrame.

#df['Valor_Venda'] > 1000: Define a condição: se o valor na coluna Valor_Venda for maior
#que 1000.
#0.15: O valor a ser atribuído à coluna Desconto' se a condição for verdadeira
 #(desconto de 15%).
#0.10: O valor a ser atribuído à coluna Desconto' se a condição for falsa (desconto de 10%).

df['Desconto'] = np.where(df['Valor_Venda'] > 1000, 0.15, 0.10)
```

```python
df.head()
```

```python
# Total por cada valor da variável
df['Desconto'].value_counts()
```

```python
print('No Total 457 Vendas Receberiam Desconto de 15%.')
```

## Pergunta de Negócio 8

### Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
# Criamos uma coluna calculando o valor de venda menos o desconto
df['Valor_Venda_Desconto'] = df['Valor_Venda'] - (df['Valor_Venda'] * df['Desconto'])
```

```python
df.head()
```

```python
# Filtrando as vendas antes do desconto de 15%
df_dsa_p8_vendas_antes_desconto = df.loc[df['Desconto'] == 0.15, 'Valor_Venda']
```

```python
# Filtrando as vendas depois do desconto de 15%
df_dsa_p8_vendas_depois_desconto = df.loc[df['Desconto'] == 0.15, 'Valor_Venda_Desconto']
```

```python
# Calcula a média das vendas antes do desconto de 15%
media_vendas_antes_desconto = df_dsa_p8_vendas_antes_desconto.mean()
```

```python
# Calcula a média das vendas depois do desconto de 15%
media_vendas_depois_desconto = df_dsa_p8_vendas_depois_desconto.mean()
```

```python
print("Média das vendas antes do desconto de 15%:", round(media_vendas_antes_desconto, 2))
```

```python
print("Média das vendas depois do desconto de 15%:", round(media_vendas_depois_desconto, 2))
```

## Pergunta de Negócio 9

### Qual o Média de Vendas Por Segmento, Por Ano e Por Mês?

Demonstre o resultado através de gráfico de linha.

```python
# criando uma colluna de mes no datframe
df['Mes'] = df['Data_Pedido'].dt.month
```

```python
df.head()
```

```python
# Fazendo o agrupamento do ano,mes,segmento e buscando os paramento do valor venda e ciado um dataframe novo
df_Media = df.groupby(['Ano', 'Mes', 'Segmento'])['Valor_Venda'].agg([np.sum, np.mean, np.median])
```

```python
# Visualizando
df_Media.head()
```

```python
# Vamos extrair os níveis e colocando em cada grupo
Anos = df_Media.index.get_level_values(0)
Mes = df_Media.index.get_level_values(1)
Segmento = df_Media.index.get_level_values(2)
```

```python
# Plot Grafico
plt.figure(figsize = (12, 6))
sns.set()
fig1 = sns.relplot(kind = 'line',
                   data = df_Media,
                   y = 'mean',
                   x = Mes,
                   hue = Segmento,
                   col = Anos,
                   col_wrap = 4)
plt.show()
```

## Pergunta de Negócio 10

### Qual o Total de Vendas Por Categoria e SubCategoria, Considerando Somente as Top 12 SubCategorias?

Demonstre tudo através de um único gráfico.

```python
df.head()
```

```python
# Agrupamos por categoria e subcategoria e calculamos a soma somente para variáveus numéricas
df_categorico = df.groupby(['Categoria',
                             'SubCategoria']).sum(numeric_only = True).sort_values('Valor_Venda',
                                                                                   ascending = False).head(12)
```

```python
# Convertemos a coluna Valor_Venda em número inteiro e classificamos por categoria
df_categorico = df_categorico[['Valor_Venda']].astype(int).sort_values(by = 'Categoria').reset_index()
```

```python
# Dataframe com categorias e subcategorias
df_categorico
```

```python
# Criamos outro dataframe somente com os totais por categoria
df_categorico2 = df_categorico.groupby('Categoria').sum(numeric_only = True).reset_index()
```

```python
# Dataframe com categorias
df_categorico2
```

```python
# Listas de cores para categorias
cores_categorias = ['#5d00de',
                    '#0ee84f',
                    '#e80e27']
```

```python
# Listas de cores para subcategorias
cores_subcategorias = ['#aa8cd4',
                       '#aa8cd5',
                       '#aa8cd6',
                       '#aa8cd7',
                       '#26c957',
                       '#26c958',
                       '#26c959',
                       '#26c960',
                       '#e65e65',
                       '#e65e66',
                       '#e65e67',
                       '#e65e68']
```

```python
# Plot

# Tamanho da figura
fig, ax = plt.subplots(figsize = (18,12))

# Gráfico das categorias
p1 = ax.pie(df_categorico2['Valor_Venda'],
            radius = 1,
            labels = df_categorico2['Categoria'],
            wedgeprops = dict(edgecolor = 'white'),
            colors = cores_categorias)

# Gráfico das subcategorias
p2 = ax.pie(df_categorico['Valor_Venda'],
            radius = 0.9,
            labels = df_categorico['SubCategoria'],
            autopct = autopct_format(df_categorico['Valor_Venda']),
            colors = cores_subcategorias,
            labeldistance = 0.7,
            wedgeprops = dict(edgecolor = 'white'),
            pctdistance = 0.53,
            rotatelabels = True)

# Limpa o centro do círculo
centre_circle = plt.Circle((0, 0), 0.6, fc = 'white')

# Labels e anotações
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(df_categorico['Valor_Venda']))), xy = (-0.2, 0))
plt.title('Total de Vendas Por Categoria e Top 12 SubCategorias')
plt.show()
```
